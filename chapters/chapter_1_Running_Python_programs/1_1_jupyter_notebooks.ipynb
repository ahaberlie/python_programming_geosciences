# Chapter 1.1 - Integrated Development Environments

In this course, we will be using [Jupyter notebooks](https://en.wikipedia.org/wiki/Project_Jupyter) as our [integrated development environment (IDE)](https://en.wikipedia.org/wiki/Integrated_development_environment). Jupyter notebooks allow us to combine [Python code](https://en.wikipedia.org/wiki/Python_(programming_language)), [Markdown code](https://en.wikipedia.org/wiki/Markdown), [program input and output](https://en.wikipedia.org/wiki/Input/output), and other useful information into one, self-contained [graphical user interface (GUI)](https://en.wikipedia.org/wiki/Graphical_user_interface).

Modern IDEs allow Python software developers like you to execute and annotate code, view graphical program output, [organize and manage software packages](https://en.wikipedia.org/wiki/Package_manager), [use software project management tools](https://en.wikipedia.org/wiki/Project_management_software), and receive feedback from tools like [syntax highlighting](https://en.wikipedia.org/wiki/Syntax_highlighting) that can analyze code to identify errors. Some IDEs even include [code completion tools](https://en.wikipedia.org/wiki/Code_completion) using [generative artificial intelligence](https://en.wikipedia.org/wiki/Generative_artificial_intelligence).

## 1.1.1 - Jupyter notebooks

Jupyter notebooks use common web browser functionality to display graphical information and connect to a [Jupyter notebook server](https://docs.jupyter.org/en/latest/running.html) that can be run locally (Fig. 1.1.f1) or remotely (e.g., <a href="https://triton.niu.edu">EAE's triton</a>, <a href="https://colab.google.com">google colab</a>, etc.). Remote servers require only a web browser to access, which greatly reduces the computing requirements on your end and eliminates the need for you to go through the complicated setup steps. Because of this, Jupyter notebooks are one of the most popular ways to develop and share code, particularly for academic research (Fig. 1.1.f2) or jobs that rely heavily on prototyping and visualization. During this course, we will be using the notebook interface to complete examples, exercises, assignments, and your final project. These notebooks should be included in a [career portfolio](https://en.wikipedia.org/wiki/Career_portfolio) that demonstrates your Python skills to future employers. In other words, the learning curve needed to master the basic usage of notebooks will be worthwhile and make you more desirable to employers. Project managers and hiring committees like to see tangible evidence of coding experience.

## 1.1.2 - Markdown Code in Jupyter notebooks

Before we learn about Python code, we will explore some examples of text formatting in Jupyter Notebooks using Markdown code. Markdown code is a [markup language](https://en.wikipedia.org/wiki/Lightweight_markup_language) that takes [plain text language (like English)](https://en.wikipedia.org/wiki/Plain_text) and formats it in specific ways using markup [syntax](https://en.wikipedia.org/wiki/Syntax_(programming_languages)). The syntax of the language defines program behavior (such as formatting text) using specific keywords and/or characters.


### 1.1.2.1 - Headings

You can organize your markdown text by using **headings**. Headings usually identify the start of a section, subsection, etc. There are several examples provided in this notebook already. If you double click on the Markdown cells, the code for those headings would look like this:

```
# Top heading
## Secondary heading
### Tertiary heading
```

The character that tells Jupyter to display a top heading or secondary heading is the ```#``` symbol, and an increasing number of these symbols correspond to lower and lower headings.

### 1.1.2.2 - Italics

*Italic text*

_Italic text_

Markdown code:

```
*Italic text*
_Italic text_
```

### 1.1.2.3 - Bold

*Bold text**

__Bold text__

Markdown code:

```
**Bold text**
__Bold text__
```

### 1.1.2.4 - Strikethrough

<s>Strikethrough text</s>

Markdown code:

```
~~Strikethrough text~~
```

### 1.1.2.5 - Code blocks

Code examples can and should be formatted like the following example. This will keep the correct spacing, text appearance, etc., that will make it look like code and not text. Please note that this code is **not** runable and is only used for explanatory purposes.

```
a = 10
b = 20
```

Markdown code:

\```

a = 10

b = 20

\```

### 1.1.2.6 - Indentation

> Indent once

>> Indent twice

>>> Indent thrice

Markdown code:

```
> Indent once

>> Indent twice

>>> Indent thrice
```

### 1.1.2.7 - Bullets and Numbering

### 1.1.2.8 - Horizontal lines

## 1.1.3 - HyperText Markup Language (HTML) 

Jupyter notebook markdown cells can use most HTML syntax. It is more difficult to use than Markdown, but it can provide additional customization if it is needed. Unlike typical HTML files, you only need to include individual [tags](https://www.w3schools.com/TAGS/default.asp). There are many available tags (sometimes known as "elements"), but since this is not an HTML course, we will only focus on a few useful tags.

### 1.1.3.1 - Line breaks

Sometimes, text will appear on the same line even if you want it to be on multiple lines. If this happens, you can try the ```<br>``` tag.

Text on one line<br>Text on second line

Markdown code:

```
Text on one line<br>Text on second line
```

### 1.1.3.2 - Text color

<font color=blue>Blue Text</font>

<font color=red>Red Text</font>

<font color=yellow>Yellow Text</font>

Markdown code:

```
<font color=blue>Blue Text</font>

<font color=red>Red Text</font>

<font color=yellow>Yellow Text</font>
```

## 1.1.4 - LaTeX and Equations

Like HTML, Markdown cells can render most [LaTeX](https://en.wikipedia.org/wiki/LaTeX) code. This is a markup language that is generally used to format documents. In Jupyter notebooks, it is often used for [mathematical equations](http://csrgxtu.github.io/2015/03/20/Writing-Mathematic-Fomulars-in-Markdown/). Learning LaTeX is beyond the scope of this course, but here are a few examples that could help you format your own equations during the course:

$ e = mc^2 $

$ f = ma $

$\infty$

$k_{n+1}$

$\frac{n!}{k!(n-k)!}$

$\sqrt{k}$

$\int\limits_a^b$

Markdown code

```
$ e = mc^2 $

$ f = ma $

$\infty$

$k_{n+1}$

$\frac{n!}{k!(n-k)!}$

$\sqrt{k}$

$\int\limits_a^b$
```
